```python
from scipy.sparse import coo_matrix
from sklearn.utils import shuffle
from sklearn.preprocessing import LabelEncoder
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer, load_iris, load_wine, load_digits


class Multi_knn_cl():
    def __init__(self, ncls, models12):
        self.ncls = ncls
        self.models12 = models12
        self.knn4 = KNeighborsClassifier(
            n_neighbors=4, leaf_size=30, p=2,  weights='distance')
        self.knn5 = KNeighborsClassifier(
            n_neighbors=5, leaf_size=30, p=2,  weights='distance')
        self.knn6 = KNeighborsClassifier(
            n_neighbors=6, leaf_size=30, p=2,  weights='distance')
        self.knn7 = KNeighborsClassifier(
            n_neighbors=7, leaf_size=30, p=2,  weights='distance')
        self.knn8 = KNeighborsClassifier(
            n_neighbors=8, leaf_size=30, p=2,  weights='distance')
        self.knn9 = KNeighborsClassifier(
            n_neighbors=9, leaf_size=30, p=2,  weights='distance')
        if self.models12:
            self.knn41 = KNeighborsClassifier(
                n_neighbors=4, leaf_size=30, p=1.5, weights='distance')
            self.knn51 = KNeighborsClassifier(
                n_neighbors=5, leaf_size=30, p=1.5, weights='distance')
            self.knn61 = KNeighborsClassifier(
                n_neighbors=6, leaf_size=30, p=1.5, weights='distance')
            self.knn71 = KNeighborsClassifier(
                n_neighbors=7, leaf_size=30, p=1.5, weights='distance')
            self.knn81 = KNeighborsClassifier(
                n_neighbors=8, leaf_size=30, p=1.5, weights='distance')
            self.knn91 = KNeighborsClassifier(
                n_neighbors=9, leaf_size=30, p=1.5, weights='distance')

    def fit(self, X, y):
        self.knn4.fit(X, y)
        self.knn5.fit(X, y)
        self.knn6.fit(X, y)
        self.knn7.fit(X, y)
        self.knn8.fit(X, y)
        self.knn9.fit(X, y)
        if self.models12:
            self.knn41.fit(X, y)
            self.knn51.fit(X, y)
            self.knn61.fit(X, y)
            self.knn71.fit(X, y)
            self.knn81.fit(X, y)
            self.knn91.fit(X, y)

    def predict(self, X_test):
        anarray = np.zeros((X_test.shape[0], self.ncls))
        ypred = self.knn4.predict_proba(X_test)
        anarray[:, :] = anarray[:, :]+ypred[:, :]
        ypred = self.knn5.predict_proba(X_test)
        anarray[:, :] = anarray[:, :]+ypred[:, :]
        ypred = self.knn6.predict_proba(X_test)
        anarray[:, :] = anarray[:, :]+ypred[:, :]
        ypred = self.knn7.predict_proba(X_test)
        anarray[:, :] = anarray[:, :]+ypred[:, :]
        ypred = self.knn8.predict_proba(X_test)
        anarray[:, :] = anarray[:, :]+ypred[:, :]
        ypred = self.knn9.predict_proba(X_test)
        anarray[:, :] = anarray[:, :]+ypred[:, :]
        if self.models12:
            ypred = self.knn41.predict_proba(X_test)
            anarray[:, :] = anarray[:, :]+ypred[:, :]
            ypred = self.knn51.predict_proba(X_test)
            anarray[:, :] = anarray[:, :]+ypred[:, :]
            ypred = self.knn61.predict_proba(X_test)
            anarray[:, :] = anarray[:, :]+ypred[:, :]
            ypred = self.knn71.predict_proba(X_test)
            anarray[:, :] = anarray[:, :]+ypred[:, :]
            ypred = self.knn81.predict_proba(X_test)
            anarray[:, :] = anarray[:, :]+ypred[:, :]
            ypred = self.knn91.predict_proba(X_test)
            anarray[:, :] = anarray[:, :]+ypred[:, :]
        if self.models12:
            z = anarray/12.
        else:
            z = anarray/6.
        labels = np.argmax(z, axis=1)
        classes = self.knn4.classes_
        labels = [classes[i] for i in labels]
        return labels

    def predict_proba(self, X_test):
        anarray = np.zeros((X_test.shape[0], self.ncls))
        ypred = self.knn4.predict_proba(X_test)
        anarray[:, :] = anarray[:, :]+ypred[:, :]
        ypred = self.knn5.predict_proba(X_test)
        anarray[:, :] = anarray[:, :]+ypred[:, :]
        ypred = self.knn6.predict_proba(X_test)
        anarray[:, :] = anarray[:, :]+ypred[:, :]
        ypred = self.knn7.predict_proba(X_test)
        anarray[:, :] = anarray[:, :]+ypred[:, :]
        ypred = self.knn8.predict_proba(X_test)
        anarray[:, :] = anarray[:, :]+ypred[:, :]
        ypred = self.knn9.predict_proba(X_test)
        anarray[:, :] = anarray[:, :]+ypred[:, :]
        if self.models12:
            ypred = self.knn41.predict_proba(X_test)
            anarray[:, :] = anarray[:, :]+ypred[:, :]
            ypred = self.knn51.predict_proba(X_test)
            anarray[:, :] = anarray[:, :]+ypred[:, :]
            ypred = self.knn61.predict_proba(X_test)
            anarray[:, :] = anarray[:, :]+ypred[:, :]
            ypred = self.knn71.predict_proba(X_test)
            anarray[:, :] = anarray[:, :]+ypred[:, :]
            ypred = self.knn81.predict_proba(X_test)
            anarray[:, :] = anarray[:, :]+ypred[:, :]
            ypred = self.knn91.predict_proba(X_test)
            anarray[:, :] = anarray[:, :]+ypred[:, :]
        if self.models12:
            z = anarray/12.
        else:
            z = anarray/6.
        return z


def hassanat_dist_metric(x1, x2):
    tmp = 0.
    for i in range(len(x1)):
        min_ = min(x1[i], x2[i])
        max_ = max(x1[i], x2[i])
        if min_ >= 0:
            di = 1-((1+min_)/(1+max_))
        else:
            di = 1-((1+min_ + np.abs(min_))/(1+max_+np.abs(min_)))
        tmp = tmp+di
    return tmp


def hassanat_dist_metric1(x1, x2):
    tmp = 0.
    for i in range(len(x1)):
        min_ = min(x1[i], x2[i])
        max_ = max(x1[i], x2[i])
        if min_ >= 0:
            di = 1-(1+min_)/(1+max_)
        else:
            di = 1-1/(1+max_+np.abs(min_))
        tmp = tmp+di
    return tmp


def hassanat_dist_metric2(x1, x2):
    tmp = 1.
    for i in range(len(x1)):
        min_ = min(x1[i], x2[i])
        max_ = max(x1[i], x2[i])
        if min_ >= 0:
            di = 1-np.abs(x1[i]-x2[i])/(1+max_)
        else:
            di = 1-np.abs(x1[i]-x2[i])/(1+max_+np.abs(min_))
        tmp = tmp-di
    return tmp


def LD(x, y):
    tmp = 0.
    for xi, yi in zip(x, y):
        tmp += np.log(1. + np.abs(xi-yi))
    return tmp


def load_letter():
    ylabel = []
    X = []
    filename = 'letter-recognition.data.csv'
    afile = open(filename, 'r')
    jline = 0
    for line in afile:
        jline = jline+1
        if jline > 1:
            X.append([int(line.split(',')[j])
                      for j in range(1, len(line.split(',')))])
            astring = line.split(',')[0]
            astring = astring.strip()
            ylabel.append(astring)
    afile.close()
    X = np.array(X)
    encoder = LabelEncoder()
    encoder.fit(ylabel)
    y = encoder.transform(ylabel)
    y = np.array(y)
    return X, y


def load_balance():
    y = []
    X = []
    filename = 'balance-scale.data.csv'
    afile = open(filename, 'r')
    jline = 0
    for line in afile:
        jline = jline+1
        if jline > 1:
            X.append([int(line.split(',')[j])
                      for j in range(1, len(line.split(',')))])
            astring = line.split(',')[0]
            astring = astring.strip()
            if astring == 'B':
                y.append(0)
            elif astring == 'R':
                y.append(1)
            else:
                y.append(2)
    afile.close()
    X = np.array(X)
    y = np.array(y)
    return X, y


def load_wisconsin():
    y = []
    X = []
    filename = 'breast-cancer-wisconsin.data.csv'
    afile = open(filename, 'r')
    jline = 0
    for line in afile:
        jline = jline+1
        if jline > 1:
            X.append([float(line.split(',')[j])
                      for j in range(2, len(line.split(',')))])
            astring = line.split(',')[1]
            astring = astring.strip()
            if astring == 'M':
                y.append(1)
            else:
                y.append(0)
    afile.close()
    X = np.array(X)
    y = np.array(y)
    return X, y


def load_sonar():
    y = []
    X = []
    filename = 'sonar.all-data.csv'
    afile = open(filename, 'r')
    for line in afile:
        X.append([float(line.split(',')[j])
                 for j in range(len(line.split(','))-1)])
        astring = line.split(',')[-1]
        astring = astring.strip()
        if astring == 'M':
            y.append(1)
        else:
            y.append(0)
    afile.close()
    X = np.array(X)
    y = np.array(y)
    return X, y


def load_haberman():
    y = []
    X = []
    filename = 'haberman.data.csv'
    afile = open(filename, 'r')
    for line in afile:
        X.append([int(line.split(',')[j])
                 for j in range(len(line.split(','))-1)])
        y.append(int(line.split(',')[-1]))
    afile.close()
    X = np.array(X)
    y = np.array(y)
    return X, y


def get_Xandy(iitt):
    if iitt == 0:
        data = load_breast_cancer()
        X = data.data
        y = data.target
    if iitt == 1:
        data = load_iris()
        X = data.data
        y = data.target
    if iitt == 2:
        data = load_wine()
        X = data.data
        y = data.target
    if iitt == 3:
        data = load_digits()
        X = data.data
        y = data.target
    if iitt == 4:
        X, y = load_haberman()
    if iitt == 5:
        X, y = load_sonar()
    if iitt == 6:
        X, y = load_wisconsin()
    if iitt == 7:
        X, y = load_balance()
    if iitt == 8:
        X, y = load_letter()
    return X, y


def get_accuracy_metric(X, y, irand):
    train_data, test_data, train_y, test_y = train_test_split(
        X, y, test_size=0.25, shuffle=True, random_state=irand)
    scaler = StandardScaler()
    train_data = scaler.fit_transform(train_data)
    test_data = scaler.transform(test_data)
    models12 = False
#    print(np.unique(y))
    ncls = len(np.unique(y))
    clf = Multi_knn_cl(ncls, models12)
    clf.fit(train_data, train_y)
    preds = clf.predict(test_data)
    return accuracy_score(test_y, preds)


def get_accuracy_metricnn(X, y, nn, irand):
    train_data, test_data, train_y, test_y = train_test_split(
        X, y, test_size=0.25, shuffle=True, random_state=irand)
    scaler = StandardScaler()
    train_data = scaler.fit_transform(train_data)
    test_data = scaler.transform(test_data)
    clf = KNeighborsClassifier(
        n_neighbors=nn, weights='distance', leaf_size=30)
    clf.fit(train_data, train_y)
    preds = clf.predict(test_data)
    return accuracy_score(test_y, preds)
```
